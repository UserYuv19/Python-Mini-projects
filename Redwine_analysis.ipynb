```python
if data['quality'].isnull().any():
    data = data.dropna(subset=['quality'])
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = data.drop('quality', axis=1)
y = data['quality']
```

```python
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
```

```python
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
```

```python
outlier_detector = IsolationForest(contamination=0.05)  # Adjust contamination based on your dataset
outliers = outlier_detector.fit_predict(X_imputed)
X_no_outliers = X_imputed[outliers != -1]
y_no_outliers = y[outliers != -1]
```

```python
estimator = RandomForestClassifier()
selector = RFE(estimator, n_features_to_select=5)
X_selected = selector.fit_transform(X_no_outliers, y_no_outliers)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_no_outliers, test_size=0.2, random_state=42)
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_selected[:, 0], X_selected[:, 4], c=y[outliers != -1], cmap='viridis', edgecolor='k')
plt.xlabel('Fixed Acidity')
plt.ylabel('Alcohol')
plt.title('Scatter Plot of Fixed Acidity vs Alcohol')
plt.colorbar(label='Wine Quality')
plt.show()
```
